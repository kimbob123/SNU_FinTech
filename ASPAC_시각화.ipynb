```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 데이터 설정
data1 = {
    'Year': [2020, 2021, 2022],
    'Size': [1468, 81625, 249648]
}

data2 = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Size': [34500, 101700, 188200, 300300, 376600]
}

data3 = {
    'Year': [2023, 2024, 2025, 2026, 2027],
    'Size': [448500, 507900, 561000, 610500, 660100]
}

# CAGR 계산 함수
def calculate_cagr(start_value, end_value, periods):
    return (end_value / start_value) ** (1 / periods) - 1

# 데이터프레임 생성
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)

# CAGR 계산
cagr1 = calculate_cagr(df1['Size'].iloc[0], df1['Size'].iloc[-1], len(df1['Year']) - 1)
cagr2 = calculate_cagr(df2['Size'].iloc[0], df2['Size'].iloc[-1], len(df2['Year']) - 1)
cagr3 = calculate_cagr(df3['Size'].iloc[0], df3['Size'].iloc[-1], len(df3['Year']) - 1)

# 시각화
sns.set(style="whitegrid")

def plot_data(df, cagr, title):
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Year', y='Size', data=df, palette=sns.dark_palette("#69d", reverse=True, as_cmap=True))
    plt.title(title, fontsize=20)
    plt.xlabel('Year')
    plt.ylabel('Market Size (in one million won)')
    plt.legend(labels=[f'CAGR: {cagr:.2%}'], loc='upper left')
    plt.show()

# 데이터1 그래프
plot_data(df1, cagr1, 'Market Size over Years')

# 데이터2 그래프
plot_data(df2, cagr2, 'Market Size over Years')

# 데이터3 그래프
plot_data(df3, cagr3, 'Market Size and Forecast')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 데이터 설정
data1 = {
    'Year': [2020, 2021, 2022],
    'Size': [14.68, 816.25, 2496.48]
}

data2 = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Size': [345, 1017, 1882, 3003, 3766]
}

data3 = {
    'Year': [2023, 2024, 2025, 2026, 2027],
    'Size': [4485, 5079, 5610, 6105, 6601]
}

# CAGR 계산 함수
def calculate_cagr(start_value, end_value, periods):
    return (end_value / start_value) ** (1 / periods) - 1

# 데이터프레임 생성
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)

# CAGR 계산
cagr1 = calculate_cagr(df1['Size'].iloc[0], df1['Size'].iloc[-1], len(df1['Year']) - 1)
cagr2 = calculate_cagr(df2['Size'].iloc[0], df2['Size'].iloc[-1], len(df2['Year']) - 1)
cagr3 = calculate_cagr(df3['Size'].iloc[0], df3['Size'].iloc[-1], len(df3['Year']) - 1)

# 시각화
sns.set(style="white")

def plot_data(df, cagr, title):
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x='Year', y='Size', hue='Year', dodge=False, data=df, palette='Blues', legend=False)
    for p in ax.patches:
        ax.annotate(format(p.get_height(), ',.0f'),
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha = 'center', va = 'center',
                    xytext = (0, 6),
                    textcoords = 'offset points')
    plt.title(title, fontsize=20)
    plt.xlabel('Year')
    plt.ylabel('Market Size (in one billion won)')
    plt.legend(labels=[f'CAGR: {cagr:.2%}'], loc='upper left')
    plt.show()

# 데이터1 그래프
plot_data(df1, cagr1, 'Meal Kits Market Size_Korea MFDS')

# 데이터2 그래프
plot_data(df2, cagr2, 'Meal Kits Market Size_Euromonitor')

# 데이터3 그래프
plot_data(df3, cagr3, 'Meal Kits Market Size and Forecast')
```
